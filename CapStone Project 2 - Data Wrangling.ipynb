```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
games = pd.read_csv('C:/Users/michael.thabane/Documents/Springboard/Capstone2/nba-games/games.csv')
games_details = pd.read_csv('C:/Users/michael.thabane/Documents/Springboard/Capstone2/nba-games/games_details.csv')
players = pd.read_csv('C:/Users/michael.thabane/Documents/Springboard/Capstone2/nba-games/players.csv')
standings = pd.read_csv('C:/Users/michael.thabane/Documents/Springboard/Capstone2/nba-games/ranking.csv')
```

```python
# convert date string to datetime variable
from datetime import datetime
games['GAME_DATE_EST'] = [datetime.strptime(x,'%Y-%m-%d') for x in games['GAME_DATE_EST']]
standings['STANDINGSDATE'] = [datetime.strptime(x,'%Y-%m-%d') for x in standings['STANDINGSDATE']]
```

```python
#create season variable
standings['SEASON'] = np.where(standings['SEASON_ID'] < 2000, standings['SEASON_ID'] - 10000, standings['SEASON_ID'] - 20000)
```

```python
games_details.info()
```

```python
final_boxscore = games_details.merge(games, left_on='GAME_ID', right_on="GAME_ID", how='inner')
```

```python
final_boxscore = final_boxscore[['SEASON','GAME_ID','GAME_DATE_EST','TEAM_ID','HOME_TEAM_ID','VISITOR_TEAM_ID','TEAM_ABBREVIATION',
                                'TEAM_CITY','PLAYER_ID','PLAYER_NAME','START_POSITION','MIN', 'PTS','FGM','FGA','FG_PCT',
                                'FG3M', 'FG3A','FG3_PCT','FTM','FTA','FT_PCT','OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TO',
                                'PF', 'PLUS_MINUS','PTS_home','PTS_away', 'AST_home','AST_away','REB_home','REB_away',
                                'FG_PCT_home','FG_PCT_away','FT_PCT_home','FT_PCT_away','FG3_PCT_home','FG3_PCT_away']]
```

```python
final_boxscore['GAME_LOC'] = np.where(final_boxscore['TEAM_ID'] == final_boxscore['HOME_TEAM_ID'], 'HOME','AWAY')
```

```python
final_boxscore['TOT_PTS'] = np.where(final_boxscore['GAME_LOC'] == 'HOME', final_boxscore['PTS_home'] , final_boxscore['PTS_away'])
final_boxscore['TOT_AST'] = np.where(final_boxscore['GAME_LOC'] == 'HOME', final_boxscore['AST_home'] , final_boxscore['AST_away'])
final_boxscore['TOT_REB'] = np.where(final_boxscore['GAME_LOC'] == 'HOME', final_boxscore['REB_home'] , final_boxscore['REB_away'])
final_boxscore['TOT_FG_PCT'] = np.where(final_boxscore['GAME_LOC'] == 'HOME', final_boxscore['FG_PCT_home'] , final_boxscore['FG_PCT_away'])
final_boxscore['TOT_FT_PCT'] = np.where(final_boxscore['GAME_LOC'] == 'HOME', final_boxscore['FT_PCT_home'] , final_boxscore['FT_PCT_away'])
final_boxscore['TOT_FG3_PCT'] = np.where(final_boxscore['GAME_LOC'] == 'HOME', final_boxscore['FG3_PCT_home'] , final_boxscore['FG3_PCT_away'])
```

```python
final_boxscore = final_boxscore[['SEASON','GAME_ID','GAME_DATE_EST','TEAM_ID','HOME_TEAM_ID','VISITOR_TEAM_ID','TEAM_ABBREVIATION',
                                'TEAM_CITY','GAME_LOC','PLAYER_ID','PLAYER_NAME','START_POSITION','MIN', 'PTS','FGM','FGA',
                                'FG_PCT','FG3M', 'FG3A','FG3_PCT','FTM','FTA','FT_PCT','OREB', 'DREB', 'REB', 'AST', 'STL', 
                                'BLK','TO', 'PF', 'PLUS_MINUS','TOT_PTS','TOT_AST','TOT_REB','TOT_FG_PCT','TOT_FT_PCT',
                                'TOT_FG3_PCT']]
```

```python
all_players = players['PLAYER_ID']
all_players = list(all_players.unique())
print(len(all_players))
```

```python
rolling = pd.DataFrame(columns=['GAME_DATE_EST','PLAYER_ID', 'PTSL5','PTSL3'])
for player in all_players:
    player_stats = final_boxscore.loc[final_boxscore['PLAYER_ID'] == player]
    player_stats.index = np.flip(np.arange(1,len(player_stats) + 1))
    player_roll = []
    for i in np.arange(1,len(player_stats) + 1):
        roll_list = [player_stats.iloc[i-1]['GAME_DATE_EST'], 
                    player_stats.iloc[i-1]['PLAYER_ID'],
                    player_stats.iloc[i-1:i+5]['PTS'].mean(),
                    player_stats.iloc[i-1:i+3]['PTS'].mean()]
        player_roll.append(roll_list)
    Var = pd.DataFrame(player_roll)
    Var.columns = ['GAME_DATE_EST','PLAYER_ID', 'PTSL5','PTSL3']
    rolling = rolling.append(Var)
```
